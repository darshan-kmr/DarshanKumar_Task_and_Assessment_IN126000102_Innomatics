```python
class light:
    def __init__(self,name):
        self.name=name
    def action(self,state):
        print(f"{self.name} is {state}")

l=light("Bedroom Light")
l.action("ON")
```

```python
class employee:
    def __init__(self,name,i,d):
        self.name=name
        self.i=i
        self.d=d
    def show(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.i}")
        print(f"Department: {self.d}")

e=employee("Rahul","EMP102","IT")
e.show()
```

```python
class contact:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def show(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

c=contact("Anita","9876543210")
c.show()
```

```python
class product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def show(self):
        print(f"Product: {self.name}")
        print(f"Price: {self.price}")

p=product("Headphones",2499)
p.show()
```

```python
class movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def show(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

m=movie("Inception",4.8)
m.show()
```

```python
class delivery:
    def __init__(self,name,address):
        self.name=name
        self.address=address
    def show(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")

d=delivery("Suman","Hyderabad")
d.show()
```
