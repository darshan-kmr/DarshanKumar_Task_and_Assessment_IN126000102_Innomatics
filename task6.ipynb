```python
def parking(cap,logs):
    c=0
    peak=0
    for l in logs:
        if l=="IN":
            c+=1
        elif l=="OUT" and c>0:
            c-=1
        if c>peak:
            peak=c
    status="full" if c>=cap else "available"
    print("currently parked vehicles:",c)
    print("parking status:",status)

parking(50,["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def eta(dist,traffic,weather):
    t=dist*5
    if traffic=="high":
        t+=15
    elif traffic=="medium":
        t+=10
    if weather=="rainy":
        t+=10
    print("estimated delivery time:",t,"minutes")

eta(8,"high","rainy")
```

```python
def theatre(total,booked):
    occ=(len(booked)/total)*100
    if occ==100:
        status="housefull"
    elif occ>=75:
        status="almost full"
    else:
        status="available"
    print("occupancy:",occ,"%")
    print("show status:",status)

theatre(200,[1]*150)
```

```python
def server(data):
    s=0
    for d in data:
        s+=d
    avg=s/len(data)
    if avg<50:
        status="normal"
    elif avg<=80:
        status="warning"
    else:
        status="critical"
    print("average cpu load:",avg,"%")
    print("server status:",status)

server([45,60,70,85,90])
```

```python
def classroom(d):
    over=[]
    for k,v in d.items():
        if v>8:
            over.append(k)
    alert="yes" if over else "no"
    print("overused resources:",",".join(over))
    print("energy alert:",alert)

classroom({"Projector":6,"AC":9,"Lights":4})
```

```python
def registration(cap,regs):
    confirmed=0
    for i in range(regs):
        if confirmed<cap:
            confirmed+=1
    wait=regs-confirmed
    status="closed" if confirmed>=cap else "open"
    print("confirmed registrations:",confirmed)
    print("waitlisted users:",wait)
    print("registration status:",status)

registration(100,105)
```
