```python
import string
s="buy mobile phone buy phone online".lower()
for p in string.punctuation:
    s=s.replace(p,"")
words=s.split()
freq={}
for w in words:
    freq[w]=freq.get(w,0)+1
print({k:v for k,v in freq.items() if v>1})
```

```python
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
d={}
for e in emails:
    domain=e.split("@")[1]
    d[domain]=d.get(domain,0)+1
t=len(emails)
for k,v in d.items():
    print(f"{k}:{(v/t)*100}%")
```

```python
sales=[1200,1500,900,2200,1400,3000]
avg=sum(sales)/len(sales)
for i,v in enumerate(sales,1):
    if v>avg*1.3:
        print("day",i,":",v)
```
